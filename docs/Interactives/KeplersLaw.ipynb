```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# 1. Define Constants
# ----------------------------
G = 6.67430e-11  # gravitational constant (m^3 kg^-1 s^-2)
M = 1.989e30     # mass of the Sun (kg)
AU = 1.496e11    # astronomical unit (m)

# ----------------------------
# 2. Setup Orbital Radii and Periods
# ----------------------------
radii = np.linspace(0.2 * AU, 5 * AU, 8)  # 8 orbits from 0.2 AU to 5 AU
colors = plt.cm.viridis(np.linspace(0, 1, len(radii)))

T_squared = (4 * np.pi**2 / (G * M)) * radii**3
T = np.sqrt(T_squared)

# ----------------------------
# 3. Visualize Orbits (Top-Down View)
# ----------------------------
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.set_title("Circular Orbits around a Star")
ax.set_xlabel("x [m]")
ax.set_ylabel("y [m]")
ax.grid(True)

# Plot each orbit
theta = np.linspace(0, 2 * np.pi, 500)
for r, color in zip(radii, colors):
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    ax.plot(x, y, color=color, linestyle='--')

    # Place "planet" at  = 0
    ax.plot(r, 0, 'o', color=color)

# Plot the central star
ax.plot(0, 0, 'yo', markersize=12, label='Star')
ax.legend()
plt.show()

# ----------------------------
# 4. Plot T^2 vs r^3
# ----------------------------
T_squared = T**2
r_cubed = radii**3

plt.figure(figsize=(8, 6))
plt.plot(r_cubed, T_squared, 'o-', color='darkorange')
plt.title("Kepler's Third Law: $T^2$ vs $r^3$")
plt.xlabel("Orbital Radius Cubed ($r^3$) [m]")
plt.ylabel("Orbital Period Squared ($T^2$) [s]")
plt.grid(True)
plt.show()
```
