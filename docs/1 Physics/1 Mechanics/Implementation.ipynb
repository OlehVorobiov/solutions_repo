```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Gravity (m/s^2)
v0 = 20   # Initial velocity (m/s)
angles = [15, 30, 45, 60, 75]  # Angles in degrees

# Function to compute trajectory
def projectile_trajectory(v0, theta):
    theta_rad = np.radians(theta)  # Convert to radians
    t_flight = (2 * v0 * np.sin(theta_rad)) / g  # Time of flight
    t = np.linspace(0, t_flight, num=100)  # Time intervals

    # Equations of motion
    x = v0 * np.cos(theta_rad) * t
    y = v0 * np.sin(theta_rad) * t - (0.5 * g * t**2)

    return x, y

# Plot trajectories for different angles
plt.figure(figsize=(10, 6))

for angle in angles:
    x, y = projectile_trajectory(v0, angle)
    plt.plot(x, y, label=f"{angle}°")

# Graph settings
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Distance (m)")
plt.title("Projectile Motion for Different Angles")
plt.legend()
plt.grid()
plt.show()
```
